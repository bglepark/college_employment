```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('GP_통합(3years)_ver4.csv')
df
```

```python
con_1 = df['취업여부']==0
con_2 = df['석사/박사 재학여부']==1
con_3 = df['석사/박사 재학여부']==2
con_4 = df['석사/박사 재학여부']==3

condition = con_1&(con_2|con_3|con_4)
```

```python
index = df.loc[condition].index
```

```python
df = df.drop(index)
```
