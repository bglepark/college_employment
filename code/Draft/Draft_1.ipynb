```python
df = pd.read_csv('./datasets/GP18__2019_컬럼명수정.csv')
df.head()
```

```python
test_df = df[['본분교','국공사립','학위과정유형','주야간','전공계열','전공_소분류_코드',
              '학교유형', '학교소재권역', '학교소재지역','성별','출생년', '출생월' ,'연령', '졸업년','졸업월',
              'G181SQ001'
]]

test_df.head()
```

```python
test_df.info()
```

```python
# test_df['전공_세분류_코드'].nunique()
test_df['전공_소분류_코드'].nunique()
```

```python
# 연령 -> 소수점 버리기(내림)
import math

age = []

for i in test_df['연령'].values:
    age.append(math.trunc(i))

test_df['연령'] = age
test_df['연령']
```

```python
test_df['연령'].dtypes
```

```python
# 전공_소분류_코드 인코딩 -> 라벨 인코더

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
test_df['전공_소분류_코드'] = encoder.fit_transform(test_df['전공_소분류_코드'])

print(test_df['전공_소분류_코드'].unique())

test_df.info()
```

```python
# G181SQ001 -> SQ1. 지난 4주 간 주로 한 일
# 1 : 일하였음 , 2 : 일시휴직 -> 두개 다 취업한 상태로 변경(1)
# 3~15 : 미취업 -> 0으로 변경

# 컬럼명 변경
test_df.rename(columns={'G181SQ001' : '취업여부'} , inplace=True)

# 변수값 변경
# test_df.취업여부.replace({'3':'0'})

for i in test_df['취업여부']:
    if 3<= i <=15:
        test_df['취업여부'].replace(i , 0 , inplace=True)
```

```python
for i in test_df['취업여부']:
    if i ==2:
        test_df['취업여부'].replace(i , 1 , inplace=True)
```

```python
test_df.취업여부.unique()
```

```python
# 취업 여부에 대해서 데이터 불균형 발생 -> SMOTE 사용 (과적합 방지)
print(test_df.취업여부.value_counts())

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

y = test_df['취업여부']
X = test_df.drop(columns=['취업여부'])

X_train , X_test , y_train, y_test = train_test_split(X, y , test_size=0.3 , random_state=0 , stratify=y)

smote = SMOTE(random_state=0)
X_train_over , y_train_over = smote.fit_sample(X_train, y_train)

print(X_train.shape)
print(X_train_over.shape)
print(y_train.shape)
print(y_train_over.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score, precision_score

rf = RandomForestClassifier(random_state=0)

# 학습
rf.fit(X_train_over, y_train_over)

# 평가
pred = rf.predict(X_test)

acc = accuracy_score(y_test, pred)
recall = recall_score(y_test, pred)
precision = precision_score(y_test, pred)
f1 = f1_score(y_test , pred)
roc = roc_auc_score(y_test, pred)

print('정확도:',rf.__class__.__name__ ,  acc)
print('재현율:',rf.__class__.__name__ ,  recall)
print('정밀도:',rf.__class__.__name__ ,  precision)
print('F1:',rf.__class__.__name__ ,  f1)
print('roc_auc_score:',rf.__class__.__name__ ,  roc)
```

```python
# GridSearchCV 하이퍼파라미터 튜닝

from sklearn.model_selection import GridSearchCV

params = {'n_estimators':[100],
         'max_depth' : [6, 8 , 10 , 12],
         'min_samples_split' : [8, 12, 20] , 
         'min_samples_leaf' : [8, 16 , 20]}

```

```python
grid_cv = GridSearchCV(rf , param_grid=params , cv=5 , n_jobs=-1)

grid_cv.fit(X_train_over , y_train_over)

print(grid_cv.__class__.__name__)
print('최적 하이퍼 파라미터 : ' , grid_cv.best_params_)
print('최고 예측 정확도 : ' , grid_cv.best_score_)
```

```python
rf_best = grid_cv.best_estimator_

pred = grid_cv.best_estimator_.predict(X_test)

acc = accuracy_score(y_test, pred)
recall = recall_score(y_test, pred)
precision = precision_score(y_test, pred)
f1 = f1_score(y_test , pred)
roc = roc_auc_score(y_test, pred)

print('정확도:',grid_cv.best_estimator_.__class__.__name__ ,  acc)
print('재현율:',grid_cv.best_estimator_.__class__.__name__ ,  recall)
print('정밀도:',grid_cv.best_estimator_.__class__.__name__ ,  precision)
print('F1:',grid_cv.best_estimator_.__class__.__name__ ,  f1)
print('roc_auc_score:',grid_cv.best_estimator_.__class__.__name__ ,  roc)
```
